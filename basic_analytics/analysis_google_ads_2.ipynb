```python
# CELDA 1: Importación de librerías
# Importamos las librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


print("Librerías importadas correctamente")
```

```python
# CELDA 2: Carga de datos
# Cargamos el conjunto de datos
df = pd.read_csv('data_google_ads.csv')
print("Datos cargados correctamente")
```

```python
# CELDA 3: Visualización de datos
# Mostrar las primeras filas del DataFrame
print("Primeras filas del DataFrame:")
df
```

```python
# CELDA 4: Dimensiones del DataFrame
# Obtener las dimensiones del DataFrame
print("Dimensiones del DataFrame:")
df.shape
```

```python
# CELDA 5: Descripción estadística
# Obtener la descripción estadística del DataFrame
print("Descripción estadística del DataFrame:") 
df.describe()
```

```python
# CELDA 6: Limpieza de datos
# Verificar duplicados en la columna 'Keyword'
duplicates = df['Keyword'].duplicated().sum()
print(f"Número de valores duplicados en la columna 'Keyword': {duplicates}")
if duplicates > 0:
    # Eliminar duplicados
    df = df.drop_duplicates(subset=['Keyword'])
    print("Duplicados eliminados.")

# Verificar valores nulos las columnas 'Competition' y 'Top of page bid (low range)'
df = df.dropna(subset=['Competition', 'Top of page bid (low range)'])
print("Valores nulos eliminados de las columnas 'Competition' y 'Top of page bid (low range)'.")
```

```python
# CELDA 7: Visualización después de la limpieza
# Mostrar las primeras filas del DataFrame limpio
print("Primeras filas del DataFrame después de la limpieza:")
df
```

```python
# CELDA 8: Histograma de 'Competition (indexed value)' por 'Competition'
# Crear un histograma de la columna 'Competition (indexed value)' agrupado por 'Competition'
ax = df.plot.hist(column=["Competition (indexed value)"], by="Competition", bins=10, figsize=(8,12))
```

```python
# CELDA 9: Gráfico de barras de 'Avg. monthly searches' por 'Competition'
# Crear un gráfico de barras de 'Avg. monthly searches' por 'Competition'
print("Gráfico de barras de 'Avg. monthly searches' por 'Competition':")
sns.barplot(data=df, x='Competition', y='Avg. monthly searches', errorbar=None, estimator=sum)
```
